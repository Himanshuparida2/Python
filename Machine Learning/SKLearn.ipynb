```python
import sklearn
```

```python
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
print(digits.data.shape) #Total Datasets
print(digits.target.shape) #Total Target Datasets
```

```python
print(digits.target_names) #All Target Names
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
image=digits.images[1263] #image of 6
plt.imshow(image,cmap='gray') #setting the color map
plt.title(digits.target[1263]) #labeling the image 
```

```python
fig,axes=plt.subplots(nrows=3,ncols=5,figsize=(5,3))
for i in zip(axes.ravel(),digits.images,digits.target):
    ax,img,tar=i
    ax.imshow(img,cmap='Blues')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(tar)
plt.tight_layout()
```
